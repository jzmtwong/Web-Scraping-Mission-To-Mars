```python
# import dependencies
from splinter import Browser
import pandas as pd
import requests
import os
import time
from bs4 import BeautifulSoup
```

```python
#create executable path
url = 'https://mars.nasa.gov/news/'
#create brower visit
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
#taking time to load, so need to sleep
time.sleep(3)
```

<h3>JPL Mars Space Images - Featured Image</h3>

```python
# add url, create variable "featured_image_url", open browser to Chrome

featured_image_url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser = Browser('chrome')
browser.visit(featured_image_url)
```

```python
#parse html with beautiful soup
html_image = browser.html

soup = BeautifulSoup(html_image, 'html.parser')

#get background image from url with style tag
featured_image_url  = soup.find('article')['style'].replace('background-image: url(','').replace(');', '')[1:-1]

#website main page url
main_url = 'https://www.jpl.nasa.gov'

#combined url to get image url
image_url = main_url + featured_image_url

#display image
image_url
```

```python
# Visit hemispheres url
hemispheres_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(hemispheres_url)

soup = bs(hemispheres_url, 'html.parser')
#taking time to load, so need to sleep
time.sleep(15)

#html browser
hemispheres_html = browser.html
soup = bs(hemispheres_html, 'html.parser')

# retreive all items that contain mars hemispheres information
items = soup.find_all('div', class_='item')

# Create empty list to store hemisphere urls 
hemisphere_image_urls = []

# Store the main url 
hemispheres_main_url = 'https://astrogeology.usgs.gov'

# Loop
for item in items: 
    
    # store title
    title = item.find('h3').text
    
    # store full image url link
    partial_img_url = item.find('a', class_='itemLink product-item')['href']
    
    # visit the image url link 
    browser.visit(hemispheres_main_url + partial_img_url)
    
    # HTML of each hemisphere information url 
    partial_img_html = browser.html
    
    # parse HTML for each url 
    soup = bs( partial_img_html, 'html.parser')
    
    # retrieve full image 
    img_url = hemispheres_main_url + soup.find('img', class_='wide-image')['src']
    
    # create dictionary for each and append to stored list 
    hemisphere_image_urls.append({"title" : title, "img_url" : img_url})
    

# display hemisphere_image_url list
hemisphere_image_url
```
